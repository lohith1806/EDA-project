**Univariate:**

1.How was the income of the big mart?

```python
#Income
#Histogram plot using matplotlib
data = df['Income'].dropna()
plt.figure(figsize=(10, 6))
ax = sns.histplot(data, bins=10, kde=False, stat='percent', color='skyblue', edgecolor='black')
for p in ax.patches:
    height = p.get_height()
    if height > 0:
        ax.text(p.get_x() + p.get_width()/2, height + 0.5, f'{height:.1f}%', ha='center', fontsize=9)

plt.title('Distribution of Ratings (with Percentages)', fontsize=14)
plt.xlabel('Income', fontsize=12)
plt.ylabel('Percentage', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

Obervation:

sales income is not to bad nor to good, so we have to see how to improve the income.

2.Which product category has more demand?

```python
#univariet
#Product category
#barplot using seaborn

counts = df['Product_Category'].value_counts(normalize=True) * 100  # percentages
labels = counts.index
values = counts.values

plt.figure(figsize=(10,6))
sns.barplot(x=labels, y=values, palette="viridis")

# Add percentages on top
for i, val in enumerate(values):
    plt.text(i, val + 0.5, f'{val:.1f}%', ha='center', fontsize=10)

plt.title('Product Category Distribution (%)')
plt.ylabel('Percentage')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

observation:

Grocery slaes are slightly higher than other categories.
